```python
# prompt: /content/부산광역시 주민등록인구통계,안심귀갓길.csv 을 읽고 y축에는 "안심귀갓길수"을. x축에는 "구군명"을 적어서 그래프로 그려줘. 안심귀갓길수가 N/A인 구는 제외하고 그려줘

!sudo apt-get install -y fonts-nanum
!sudo fc-cache -fv
!rm ~/.cache/matplotlib -rf

import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')

try:
    df = pd.read_csv('/content/부산광역시 주민등록인구통계,안심귀갓길.csv', encoding='cp949') # 파일 인코딩 지정
    df = df.dropna(subset=['안심귀갓길수']) # 안심귀갓길수가 N/A인 행 제거

    # '안심귀갓길수' 열을 숫자형으로 변환
    df['안심귀갓길수'] = pd.to_numeric(df['안심귀갓길수'], errors='coerce')
    df = df.dropna(subset=['안심귀갓길수'])

    plt.figure(figsize=(12, 6))  # 그래프 크기 조정
    plt.bar(df['구군명'], df['안심귀갓길수'])
    plt.xlabel('구군명')
    plt.ylabel('안심귀갓길수')
    plt.title('구군별 안심귀갓길 수')
    plt.xticks(rotation=45, ha='right') # x축 레이블 회전
    plt.tight_layout() # 레이아웃 조정
    plt.show()
except FileNotFoundError:
    print("파일을 찾을 수 없습니다. 파일 경로를 확인해주세요.")
except Exception as e:
    print(f"오류 발생: {e}")
```

```python
# prompt: /content/경찰청 부산광역시경찰청_경찰서별 5대 범죄 발생 현황_20231231.csv 와/content/부산광역시 주민등록인구통계,안심귀갓길.csv 를 읽어줘 경찰서와 구군명을 매치 시켜줄게. 중구-중부,동구-동부   구군명별로 이제 발생한 성범죄 대비 안심귀갓길을 "SCR"로 정의하고 Y축에 SCR, X축에 "경찰서"을 적어서 그래프로 그려줘

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# 폰트 설정
!sudo apt-get install -y fonts-nanum
!sudo fc-cache -fv
!rm ~/.cache/matplotlib -rf
import matplotlib.font_manager as fm
fm.fontManager.addfont('/usr/share/fonts/truetype/nanum/NanumGothic.ttf')
plt.rc('font', family='NanumBarunGothic')

# 경찰서-구군 매핑 딕셔너리
police_to_gu = {
    '중부경찰서': '중구',
    '동부경찰서': '동구',
    '서부경찰서': '서구',
    '영도경찰서': '영도구',
    '부산진경찰서': '부산진구',
    '동래경찰서': '동래구',
    '남부경찰서': '남구',
    '북부경찰서': '북구',
    '해운대경찰서': '해운대구',
    '사하경찰서': '사하구',
    '금정경찰서': '금정구',
    '강서경찰서': '강서구',
    '연제경찰서': '연제구',
    '수영경찰서': '수영구',
    '사상경찰서': '사상구',
    '기장경찰서': '기장군'
}

try:
    # 범죄 데이터 로드
    crime_df = pd.read_csv('/content/경찰청 부산광역시경찰청_경찰서별 5대 범죄 발생 현황_20231231.csv', encoding='cp949')
    crime_df = crime_df.rename(columns={'경찰서': '경찰서명'})

    # 안심귀갓길 데이터 로드
    safe_df = pd.read_csv('/content/부산광역시 주민등록인구통계,안심귀갓길.csv', encoding='cp949')
    safe_df = safe_df.dropna(subset=['안심귀갓길수'])
    safe_df['안심귀갓길수'] = pd.to_numeric(safe_df['안심귀갓길수'], errors='coerce')
    safe_df = safe_df.dropna(subset=['안심귀갓길수'])

    # 데이터 병합 및 전처리
    merged_df = pd.DataFrame()
    merged_df['경찰서'] = crime_df['경찰서명'].map(police_to_gu)
    merged_df['성범죄'] = crime_df[' 성범죄']

    merged_df = pd.merge(merged_df, safe_df, left_on='경찰서', right_on='구군명', how='inner')
    merged_df['SCR'] = merged_df['성범죄'] / merged_df['안심귀갓길수']

    # 그래프 그리기
    plt.figure(figsize=(12, 6))
    plt.bar(merged_df['경찰서'], merged_df['SCR'])
    plt.xlabel('경찰서')
    plt.ylabel('SCR (성범죄 발생 건수 / 안심귀갓길 수)')
    plt.title('구군별 성범죄 발생 건수 대비 안심귀갓길 수')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

except FileNotFoundError:
    print("파일을 찾을 수 없습니다. 파일 경로를 확인해주세요.")
except KeyError as e:
    print(f"필요한 열이 없습니다: {e}")
except Exception as e:
    print(f"오류 발생: {e}")
```
